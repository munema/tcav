## 半円

```python
SIZE = 200
im = Image.new('RGB', (SIZE, SIZE), (0, 0, 0))
draw = ImageDraw.Draw(im)
l = 0.8
l = 1-l
h = l
dx =0
dy =0.1
w = 8
draw.arc((l*SIZE + dx*SIZE, h*SIZE + dy*SIZE, SIZE*(1-l) + dx*SIZE, (1-h)*SIZE + dy*SIZE), start=90, end=270, fill=(255, 255, 255), width=w)

plt.imshow(im)
plt.show()
```

```python
def get_half_circle_img(leng = 0.8, start = 0,dx = 0, dy = 0, w = 8,SIZE = 200):
    im = Image.new('RGB', (SIZE, SIZE), (0, 0, 0))
    draw = ImageDraw.Draw(im)
    l = leng
    l = 1-l
    h = l
    st = start
    ed = st + 180
    if ed > 360:
        ed -= 360
    draw.arc((l*SIZE + dx*SIZE, h*SIZE + dy*SIZE, SIZE*(1-l) + dx*SIZE, (1-h)*SIZE + dy*SIZE), start=st, end=ed, fill=(255, 255, 255), width=w)
    return im
get_half_circle_img(start = 230)
```

## 尖り

```python
SIZE = 200
im = Image.new('RGB', (SIZE, SIZE), (0, 0, 0))
draw = ImageDraw.Draw(im)
l = 0.3
h = 0.2
dx =0
dy =0.1
w = 8
draw.line(((-l*SIZE + SIZE//2,- h*SIZE + SIZE//2), (l*SIZE + SIZE//2, SIZE//2), (-l*SIZE + SIZE//2, h*SIZE + SIZE//2)), fill=(255, 255, 255),width=w)
plt.imshow(im)
plt.show()
```

```python
def get_sharp_img(leng = 0.1, sharp = 0.2,dx = 0, dy = 0, w = 8,SIZE = 200):
    im = Image.new('RGB', (SIZE, SIZE), (0, 0, 0))
    draw = ImageDraw.Draw(im)
    l = leng
    h = sharp
    draw.line(((-l*SIZE + SIZE//2,- h*SIZE + SIZE//2), (l*SIZE + SIZE//2, SIZE//2), (-l*SIZE + SIZE//2, h*SIZE + SIZE//2)), fill=(255, 255, 255),width=w)
    return im
get_sharp_img()
```

```python
def get_concat_h(im1, im2):
    dst = Image.new('RGB', (im1.width + im2.width, im1.height))
    dst.paste(im1, (0, 0))
    dst.paste(im2, (im1.width, 0))
    return dst

def get_concat_v(im1, im2):
    dst = Image.new('RGB', (im1.width, im1.height + im2.height))
    dst.paste(im1, (0, 0))
    dst.paste(im2, (0, im1.height))
    return dst

def get_4_concat(im1, im2, im3, im4):
    dst = Image.new('RGB', (im1.width + im1.width, im1.height + im1.height))
    dst.paste(im1, (0, 0))
    dst.paste(im2, (0, im1.height))
    dst.paste(im3, (im1.width, 0))
    dst.paste(im4, (im1.width, im1.height))
    return dst.resize((im1.width,im1.height))
```
